```python
# Loading in required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
df = pd.read_csv('nobel.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Frequent gender and country
top_gender = df['sex'].value_counts().idxmax()
top_country = df['birth_country'].value_counts().idxmax()


print(f"The most frequent gender is '{top_gender}' and most frequent country is '{top_country}'.")
```

```python
# Filter for laureates born in the United States of America
df['decade'] = (df['year'] // 10) * 10  
us_counts = df[df['birth_country'] == 'United States of America'].groupby('decade').size()
total_counts = df.groupby('decade').size()

# Calculate the ratio of US-born laureates per decade
ratios = (us_counts / total_counts).fillna(0)

# Identify the decade with the highest ratio
max_decade_usa = ratios.idxmax()
highest_ratio_value = ratios.max()

# Display the result
print(f"The decade with the highest ratio of US-born laureates is {max_decade_usa}, with a ratio of {highest_ratio_value:.2f}.")
```

```python
# Group by decade, category, and sex to calculate counts
grouped = df.groupby(['decade', 'category', 'sex']).size().reset_index(name='count')

# Calculate total counts per decade and category
total_per_group = df.groupby(['decade', 'category']).size().reset_index(name='total')

# Merge the grouped data with total counts
merged = grouped.merge(total_per_group, on=['decade', 'category'])

# Calculate the proportion of each sex within each decade and category
merged['proportion'] = merged['count'] / merged['total']

# Filter for female laureates and find the row with the highest proportion
highest_proportion_row = (
    merged[merged['sex'] == 'Female']
    .sort_values(by='proportion', ascending=False)
    .iloc[0]
)

# Display the result
print("The decade and category with the highest proportion of female laureates:")
print(highest_proportion_row[['sex', 'category', 'proportion', 'decade']])
```

```python
# Group data by decade, sex, category, year, and full_name
total_gender_first = (
    df.groupby(['decade', 'sex', 'category', 'year', 'full_name'])
    .size()
    .reset_index(name='count')
)

# Filter for female laureates
female_df_all_first = total_gender_first[total_gender_first['sex'] == 'Female']

# Sort female laureates by year in ascending order
female_df_all_first = female_df_all_first.sort_values(by='year', ascending=True)

# Get the first female laureate's name and category
first_woman_name = female_df_all_first.iloc[0]['full_name']
first_woman_category = female_df_all_first.iloc[0]['category']

# Display the sorted DataFrame and first laureate information
print(f"The first woman laureate was {first_woman_name}, awarded in the category of {first_woman_category}.")
```

```python
# Group by laureate type and full name to count occurrences
indiv_org = (
    df.groupby(['laureate_type', 'full_name'])
    .size()
    .reset_index(name='count')
)

# Filter for individuals or organizations that appear more than once
repeat_list = indiv_org[indiv_org['count'] > 1]['full_name'].tolist()

# Display the result
repeat_list
```

```python
# Group by category and count the number of prizes
category_counts = df.groupby('category').size().reset_index(name='prize_count')

# Sort by the number of prizes in descending order
category_counts = category_counts.sort_values(by='prize_count', ascending=False)

# Display the result
print(category_counts)
```

```python
# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    category_counts['prize_count'],
    labels=category_counts['category'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Distribution of Nobel Prizes by Category')
plt.tight_layout()
plt.show()
```

```python
import squarify

# Prepare data for the treemap with the top 30 birthplaces
top_30_birthplace_counts = df['birth_country'].value_counts().head(30).reset_index()
top_30_birthplace_counts.columns = ['birth_country', 'count']

# Truncate labels if they are too long
top_30_birthplace_counts['label'] = top_30_birthplace_counts['birth_country'].apply(
    lambda x: x if len(x) <= 15 else x[:12] + '...'
)

# Create the treemap
plt.figure(figsize=(12, 8))
squarify.plot(
    sizes=top_30_birthplace_counts['count'],
    label=top_30_birthplace_counts['label'],
    alpha=0.8,
    text_kwargs={'fontsize': 10}  # Adjust font size dynamically
)
plt.title('Treemap of Nobel Winners by Birthplace (Top 30)')
plt.axis('off')  # Hide axes
plt.tight_layout()
plt.show()
```
